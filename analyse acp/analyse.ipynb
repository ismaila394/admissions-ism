```python
import pandas as pd
from fanalysis.pca import PCA
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

df = pd.read_csv("train.csv")
print(df.head())


num_df = df.select_dtypes(include=['float64', 'int64'])


num_df = num_df.fillna(num_df.mean())


num_df = num_df.loc[:, num_df.nunique() > 1]


scaler = StandardScaler()
X = scaler.fit_transform(num_df)


my_pca = PCA(std_unit=True)  
print(my_pca.fit(X))
print(my_pca.eig_)
plt.title("PCA Eigenvalues") 
plt.xlabel("Component Number")
plt.ylabel("Eigenvalue")
print(my_pca.plot_eigenvalues())
print(my_pca.plot_eigenvalues(type='cumulative'))
print(my_pca.plot_eigenvalues(type='percentage'))
```
