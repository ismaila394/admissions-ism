```python
import pandas as pd
from fanalysis.pca import PCA
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

df = pd.read_csv("train.csv")
print(df.head())


num_df = df.select_dtypes(include=['float64', 'int64'])


num_df = num_df.fillna(num_df.mean())


num_df = num_df.loc[:, num_df.nunique() > 1]


scaler = StandardScaler()
X = scaler.fit_transform(num_df)


my_pca = PCA(std_unit=False)  
print(my_pca.fit(X))
print(my_pca.eig_)
print(my_pca.plot_eigenvalues())
print(my_pca.plot_eigenvalues(type='cumulative'))
print(my_pca.plot_eigenvalues(type='percentage'))
var=my_pca.col_topandas()
print(var.head())
var2=my_pca.row_topandas()
print(var2.head())
print(my_pca.correlation_circle(num_x_axis=0, num_y_axis=1))
print(my_pca.plot_col_contrib(num_axis=1))
print(my_pca.mapping_row(num_x_axis=0, num_y_axis=1))


plt.show()
```
